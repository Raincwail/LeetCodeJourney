# Problem

Given a string `s`, find the length of the longest 
substring without repeating characters.

# Example 1:
```
Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3.
```
# Example 2:
```
Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
```
# Example 3:
```
Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
```

# Constraints:
```
0 <= s.length <= 5 * 104
s consists of English letters, digits, symbols and spaces.
```

### `lengthOfLongestSubstring` is the most obvious solution.

We just check the all of the substrings with unique characters and get their length iteratively.

### Example:
`"abcabcbb"`

1st "inside" iteration would be finding the "abca", and save the length of "abc", which is 3.

2nd would be "bcab", 3rd - "cabc" and so on.


### `lengthOfLongestSubstring2` is the 2nd solution that uses two pointers.

We start with two pointers at the begining, moving the second one by default until we meet a non-unique element.

In that case, we start to move the second pointer so that we have a unique substring again.

This implementation is linear in complexity and constant in space, but it can be improved by changing the mechanism of moving the fst pointer.

We try to create a unique substring using just two pointers like this:

1 iter
```
    a b c a b c b b
    |1
    |2
```

2 iter
```
    a b c a b c b b
    |1
      |2
```
...

4 iter
```
    a b c a b c b b
    |1
          |2

    a b c a b c b b
      |1
          |2
```

5 iter
```
    a b c a b c b b
      |1
            |2

    a b c a b c b b
        |1
            |2
```
...

7 iter
```
    a b c a b c b b
          |1
                |2

    a b c a b c b b
            |1
                |2

    a b c a b c b b
              |1
                |2
```
...

### `lengthOfLongestSubstring3` is the 3rd solution that uses two pointers.

It is basically the same as the second one, but a couple of tweaks were added.
In particular, now fst changes instantly without a while cycle.
